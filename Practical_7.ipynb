### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.


```python
# Exercise 9

x = np.linspace(0, 10, 100)

y1 = x ** 2  
y2 = x ** 3  

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Top subplot (y1 = x^2)
ax1.plot(x, y1, color='c', linestyle='-', label='y = x')
ax1.set_title('Quadratic Function (y = x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

# Bottom subplot (y2 = x^3)
ax2.plot(x, y2, color='g', linestyle='--', label='y = x')
ax2.set_title('Cubic Function (y = x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Exercise 21

# Create grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute z values for the function z = sin(x) + cos(y)
Z = np.sin(X) + np.cos(Y)

# Create contour plot
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(label='z = sin(x) + cos(y)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of z = sin(x) + cos(y)')

# Display the plot
plt.show()
```

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
# Exercise 26

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute z values for the function z = x^2 + y^2
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add colorbar
fig.colorbar(surf, ax=ax, label='z = x + y')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot of z = x + y')

# Display the plot
plt.show()
```

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.


```python
# Exercise 27
from math import pi
variables = ['Var1', 'Var2', 'Var3', 'Var4', 'Var5']
values = np.random.uniform(0, 1, 5)  # Random values between 0 and 1
values = np.append(values, values[0])  # Close the loop for radar chart

# Compute angle for each variable
angles = [n / float(len(variables)) * 2 * pi for n in range(len(variables))]
angles += angles[:1]  # Close the loop

# Initialize radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot data
ax.fill(angles, values, color='blue', alpha=0.4)
ax.plot(angles, values, color='blue', linewidth=2)

# Add variable labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(variables)

# Add title
ax.set_title('Radar Chart with 5 Variables', size=15, color='black', pad=20)

# Display the plot
plt.show()
```

### Exercise 34:
Create a plot with twin y-axes where one axis represents `y = x` and the other represents `y = x^2`.


```python
# Exercise 34
x = np.linspace(-5, 5, 100)

# Calculate y values for y = x and y = x^2
y_linear = x
y_quadratic = x**2

# Create figure and first axis
fig, ax1 = plt.subplots()

# Plot y = x on the first y-axis
ax1.plot(x, y_linear, 'b-', label='y = x')
ax1.set_xlabel('x')
ax1.set_ylabel('y = x', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot y = x^2 on the second y-axis
ax2.plot(x, y_quadratic, 'r-', label='y = x')
ax2.set_ylabel('y = x', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title
plt.title('Plot with Twin Y-Axes: y = x and y = x')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the plot
plt.show()
```

### Exercise 42:
Create a wind rose chart using random wind direction and speed data.


```python
# Exercise 42
from math import pi

# Generate random wind data
np.random.seed(42)  # For reproducibility
n_samples = 1000
wind_directions = np.random.uniform(0, 360, n_samples)  # Directions in degrees
wind_speeds = np.random.uniform(0, 20, n_samples)  # Speeds in m/s

# Define bins for directions
bins = 16  # Number of direction bins (e.g., 16 for 22.5-degree intervals)
direction_bins = np.linspace(0, 360, bins + 1, endpoint=True)
speed_bins = [0, 5, 10, 15, 20]  # Speed bins in m/s
speed_labels = ['0-5', '5-10', '10-15', '15-20']

# Compute histogram for each speed bin
hist_data = []
for i in range(len(speed_bins) - 1):
    mask = (wind_speeds >= speed_bins[i]) & (wind_speeds < speed_bins[i + 1])
    hist, _ = np.histogram(wind_directions[mask], bins=direction_bins, density=True)
    hist_data.append(hist * 100)  # Convert to percentage

# Convert to radians for polar plot
theta = np.linspace(0, 2 * pi, bins, endpoint=False)

# Create wind rose (polar bar plot)
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.set_theta_direction(-1)  # Clockwise
ax.set_theta_offset(pi / 2)  # North at top

# Plot stacked bars for each speed bin
bottom = np.zeros(bins)
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
for i, hist in enumerate(hist_data):
    ax.bar(theta, hist, width=2 * pi / bins, bottom=bottom, color=colors[i], alpha=0.7, label=speed_labels[i])
    bottom += hist

# Customize plot
ax.set_xticks(np.linspace(0, 2 * pi, 8, endpoint=False))
ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])
ax.set_title('Wind Rose Chart (Random Data)')
ax.legend(title='Speed (m/s)', loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Display the plot
plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
categories = ['Start', 'Revenue', 'Costs', 'Taxes', 'Profit']
values = [1000, 500, -300, -150, 0]  # Last value is a placeholder for final total
colors = ['#66B2FF', '#99FF99', '#FF9999', '#FF9999', '#66B2FF']  # Colors for bars

# Calculate cumulative values and changes
cumulative = np.cumsum([0] + values[:-1])  # Cumulative sum starting from 0
values[-1] = cumulative[-1]  # Set final value as the cumulative total
bar_heights = [values[0]] + [values[i] if i > 0 else 0 for i in range(1, len(values))]  # Heights for bars

# Create figure and axis
fig, ax = plt.subplots()

# Plot bars
for i in range(len(categories)):
    if i == 0 or i == len(categories) - 1:
        # Start and Profit bars start from y=0
        ax.bar(categories[i], bar_heights[i], bottom=0, color=colors[i], edgecolor='black')
    else:
        # Intermediate bars start from previous cumulative value
        ax.bar(categories[i], bar_heights[i], bottom=cumulative[i-1], color=colors[i], edgecolor='black')

# Add connecting lines between bars
for i in range(len(categories) - 1):
    ax.plot([i, i + 1], [cumulative[i], cumulative[i]], 'k--', linewidth=1)

# Add annotations for values
for i in range(len(categories)):
    if bar_heights[i] >= 0:
        ax.text(i, cumulative[i] + 10, f'{int(bar_heights[i])}', ha='center', va='bottom')
    else:
        ax.text(i, cumulative[i] - 10, f'{int(bar_heights[i])}', ha='center', va='top')

# Add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Value')
ax.set_title('Waterfall Chart of Financial Changes')

# Display the plot
plt.show()
```

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# Exercise 49
# Define attributes and strength values
attributes = ['Speed', 'Strength', 'Agility', 'Intelligence', 'Endurance']
values = [8, 6, 9, 7, 5]  # Manually chosen values between 0 and 10
values = np.append(values, values[0])  # Close the loop for radar chart

# Compute angles for each attribute
n = len(attributes)
angles = [n / float(n) * 2 * pi for n in range(n)]
angles += angles[:1]  # Close the loop

# Initialize radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot data
ax.fill(angles, values, color='#FF9999', alpha=0.4)
ax.plot(angles, values, color='#FF9999', linewidth=2, label='Attributes')

# Add attribute labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(attributes)

# Add y-axis labels (strength scale)
ax.set_yticks([2, 4, 6, 8, 10])
ax.set_yticklabels(['2', '4', '6', '8', '10'])

# Add title and legend
ax.set_title('Radar Chart of Attribute Strengths', size=15, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Display the plot
plt.show()
```
